```python
import json
import os
import json
import os

def extract_json_data(folder_path):
    class_level_params = []
    method_level_params = {}

    # Iterate through the provided list of JSON files
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        file = file_path
        #for file in json_files:
        if os.path.exists(file) and filename.endswith('.json'):
            print(f"extracting {file}")
            with open(file, 'r') as f:
                    data = json.load(f)
    
                    # Extract classLevelParams
                    class_level_params.extend(data.get("classLevelParams", []))
    
                    # Extract methodLevelParams and its values
                    for method, params in data.get("methodLevelParams", {}).items():
                        if method not in method_level_params:
                            method_level_params[method] = []
                        # Extend the list with values from the current method's parameters
                        method_level_params[method].extend(params)

    return class_level_params, method_level_params


# List of JSON files to read
folder_path = "./ctest4j_apache_flink_saved_mapping"  # replace with actual file paths

class_level_params, method_level_params = extract_json_data(folder_path)

print("Class Level Params:", class_level_params)
print("Method Level Params:")

# To print out the method-level params with each method's list of values
configs = []
tests = []
for method, params in method_level_params.items():
    print(f"{method}: {params}")
    tests.append(method)
    configs.extend(params)
    
configs = list(set(configs))  
tests = list(set(tests))  
print(configs)
print(f"Number of Configurations :{len(configs)}")
print(f"Number of Tests Executed :{len(tests)}")
```
